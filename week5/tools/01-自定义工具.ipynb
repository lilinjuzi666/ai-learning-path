### 1. 使用@tool装饰器定义工具

```python
#例子1：
from langchain_core.tools import tool

@tool 
def addnumber(a:int,b:int) -> int:
    """计算两个整数的和"""
    return a+b

print(f"name = {addnumber.name}")
print(f"description = {addnumber.description}")
print(f"args = {addnumber.args}")
print(f"return_direct = {addnumber.return_direct}")
```

```python
#例子2：显性
from langchain_core.tools import tool

@tool(name_or_callable="add_number",
    #   args_schema={"a":int,"b":int},
      description="add two numbers",
      return_direct = True
      ) 
def addnumber(a:int,b:int) -> int:
    """计算两个整数的和"""
    return a+b

print(f"name = {addnumber.name}")
print(f"description = {addnumber.description}")
print(f"args = {addnumber.args}")
print(f"return_direct = {addnumber.return_direct}")
```

```python
#调用工具
addnumber.invoke({"a":1,"b":2})
```

```python
#例子3：举例修改args参数的描述
from langchain_core.tools import tool
from pydantic import BaseModel, Field

class FieldInfo(BaseModel):
    a:int = Field(description="第一个整数参数")
    b:int = Field(description="第二个整数参数")

@tool(name_or_callable="add_number",
      args_schema=FieldInfo,
      description="add two numbers",
      return_direct = True
      ) 
def addnumber(a:int,b:int) -> int:
    """计算两个整数的和"""
    return a+b

print(f"name = {addnumber.name}")
print(f"description = {addnumber.description}")
print(f"args = {addnumber.args}")
print(f"return_direct = {addnumber.return_direct}")
```

### 2. StructuredTool 的 from_function()的使用


```python
#例子1：
from langchain_core.tools import StructuredTool


#声明一个函数
def search_google (query: str) -> str:
    """Search Google for the given query."""
    return "搜索结果"

#定义一个工具,将一个函数封装为一个LangChain可识别的标准化工具
search_tool = StructuredTool.from_function(
    func = search_google,
    name = "search_google",
    description = "用谷歌搜索引擎查询,返回搜索结果",
    return_direct= True #默认为False 
)

print(f"name = {search_tool.name}")
print(f"description = {search_tool.description}")
print(f"args = {search_tool.args}")
print(f"return_direct = {search_tool.return_direct}")

```

```python
search_tool.invoke({"query":"python"})
```

```python
#例子2
from langchain_core.tools import StructuredTool

class FieldInfo(BaseModel):
    query:str = Field(description="要检索的关键词")


#声明一个函数
def search_google (query: str) -> str:
    """Search Google for the given query."""
    return "搜索结果"

#定义一个工具,将一个函数封装为一个LangChain可识别的标准化工具
search_tool = StructuredTool.from_function(
    func = search_google,
    name = "search_google",
    description = "用谷歌搜索引擎查询,返回搜索结果",
    return_direct= True ,#默认为False 
    args_schema= FieldInfo
)

print(f"name = {search_tool.name}")
print(f"description = {search_tool.description}")
print(f"args = {search_tool.args}")
print(f"return_direct = {search_tool.return_direct}")

```
