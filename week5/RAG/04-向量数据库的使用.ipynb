1.数据的存储

例1 ：从TXT文档中加载数据，向量化后存储到Chroma数据库中

```python
from langchain_community.document_loaders import TextLoader
from langchain_chroma import Chroma
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import DashScopeEmbeddings
import os
import dotenv


#步骤1：创建一个TextLoader的实例，并将指定的文档加载
loader = TextLoader(
    file_path = "./asset/1.txt",
    encoding = "utf-8"
)

docs = loader.load()

#步骤2：创建文本拆分器，并拆分文档
text_splitter = CharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap = 100,   
)
splitter_docs = text_splitter.split_documents(docs)

#步骤三，创建嵌入模型
dotenv.load_dotenv()
os.environ['DASHSCOPE_API_KEY'] = os.getenv('DASHSCOPE_API_KEY')
embedding_model = DashScopeEmbeddings(
    model='text-embedding-v3',
)

#步骤四，将文档及嵌入模型传入到Chroma相关结构中，进行数据的存储
db = Chroma.from_documents(
    documents=splitter_docs,
    embedding=embedding_model,
)#没有显式指明存储位置，将数据存在了内存中，并缓存起来
```

```python
db1 = Chroma.from_documents(
    documents=splitter_docs,
    embedding=embedding_model,
    persist_directory='./asset/chroma-1',
)#持久化处理
```

```python
#需要明确，在向量数据库中，不仅存储了数据的向量，而且还存储了数据本身
#演示一下：检索的需求
```

```python
query = "什么是机器学习"

docs = db.similarity_search(query)
print(docs[0].page_content)
```

例2：操作csv文档并向量化

```python
from langchain_community.document_loaders import CSVLoader
from langchain_chroma import Chroma
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import DashScopeEmbeddings
import os
import dotenv


#步骤1：创建一个TextLoader的实例，并将指定的文档加载
loader = CSVLoader(
    file_path = "./asset/3.csv.",
    encoding = "utf-8"
)

pages = loader.load_and_split()

#步骤2：创建文本拆分器，并拆分文档
text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    chunk_size = 500,       
)
docs = text_splitter.split_documents(docs)
#步骤三，创建嵌入模型
dotenv.load_dotenv()
os.environ['DASHSCOPE_API_KEY'] = os.getenv('DASHSCOPE_API_KEY')
embedding_model = DashScopeEmbeddings(
    model='text-embedding-v3',
)

#步骤四，将文档及嵌入模型传入到Chroma相关结构中，进行数据的存储
db2 = Chroma.from_documents(
    documents=splitter_docs,
    embedding=embedding_model,
    persist_directory='./asset/chroma-2',
)
```

相似性的检索

```python
query = "哺乳动物"
docs = db.similarity_search(
    query,
    k=3
    )
print(f"查询:{query}的结果:")
for i,doc in enumerate(docs,1):#指定从1开始
    print(f"\n结果{i}:")
    print(f"内容：{doc.page_content}")
    print(f"元数据:{doc.metadata}")
```

直接对问题向量查询

```python
query = "哺乳动物"
embedding_vector = embedding.embed_query(query)

docs = db.similarity_search_by_vector(embedding_vector, k=3)

print(f"查询:'{query}' 的结果:")
for i, doc in enumerate(docs, 1):
    print(f"\n结果 {i}:")
    print(f"内容:{doc.page_content}")
    print(f"元数据:{doc.metadata}")
```

相似性检索，支持过滤元数据（filter）

```python
query = "哺乳动物"
docs = db.similarity_search(
    query,
    k=3,
    filter={"source": "动物"}
    )
print(f"查询:{query}的结果:")
for i,doc in enumerate(docs,1):#指定从1开始
    print(f"\n结果{i}:")
    print(f"内容：{doc.page_content}")
    print(f"元数据:{doc.metadata}")
```

通过L2距离(欧氏距离)分数进行搜索(similarity_search_with_score)

```python
docs = db.similarity_search_with_score(
    "量子力学是什么"
)
for doc,score in docs:
    print(f"内容:[L2距离得分={score:.3f}]{doc.page_content}[{doc.metadata}]")
```

通过余弦相似度的分数进行搜索(similarity_search_with_relevance_scores)


```python
docs = db.similarity_search_with_relevance_scores(
    "量子力学是什么"
)
for doc,score in docs:
    print(f"内容:[余弦相似度得分={score:.3f}]{doc.page_content}[{doc.metadata}]")
```

MMR（最大边际相关性，max_marginal_relevance_search）

```python
# MMR是一种平衡相关性和多样性的检索策略,避免返回高度相似的冗余结果。

docs = db.max_marginal_relevance_search(
    query="量子力学是什么",
    lambda_mult=0.8,  # 范围0-1，0为最大多样性，1为最大相关性
)

print("🔍 关于【量子力学是什么】的搜索结果:")
print("=" * 50)
for i, doc in enumerate(docs):
    print(f"\n📋 结果 {i+1}:")
    print(f"★ 内容:{doc.page_content}")
    print(f"🏷️ 标签:{', '.join(f'{k}={v}' for k, v in doc.metadata.items())}")
```
