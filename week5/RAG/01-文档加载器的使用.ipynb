```python
#1.加载txt

from langchain_community.document_loaders import TextLoader

#读取文件路径
file_path = './asset/1.txt'

#创建TextLoader对象
txt_loader = TextLoader(
    file_path=file_path,
    encoding='utf-8'
)

#加载文档
docs = txt_loader.load()

#打印文档内容
print(docs[0])
# print(len(docs))
```

```python
#加载pdf
from langchain_community.document_loaders import PyPDFLoader
pdf_loader = PyPDFLoader(
    file_path='./asset/twgx.pdf'
)
docs = pdf_loader.load()
print(docs)
print(len(docs))
```

```python
#加载网络中的pdf
from langchain_community.document_loaders import PyPDFLoader
pdf_loader = PyPDFLoader(
    file_path='https://arxiv.org/pdf/2302.03803'
)
docs = pdf_loader.load()
print(docs)
print(len(docs))
```

```python
for doc in docs:
    print(doc)
```

```python
#3.加载csv
from langchain_community.document_loaders import CSVLoader
csv_loader = CSVLoader(
    file_path='./asset/3.csv',
    encoding='utf-8',
    
)
docs = csv_loader.load()
print(len(docs))
for doc in docs:
    print(doc)
```

```python
#4.JSON
#例1：加载文件中的所有数据
from langchain_community.document_loaders import JSONLoader
json_loader = JSONLoader(
    file_path='./asset/4.json',
    jq_schema=".",#表示加载所有数据
    text_content=False,#将加载的json对象转换为json字符串
)
docs = json_loader.load()
print(len(docs))
print(docs[0])
```

```python
#例2加载json中edcuation[]中的school字段
from langchain_community.document_loaders import JSONLoader
json_loader = JSONLoader(
    # file_path='./asset/4.json',
    # jq_schema=".education[].school",#表示加载特定数据
    # text_content=False,#将加载的json对象转换为json字符串
    file_path='./asset/4.json',
    jq_schema=".education[]",#表示加载特定数据
    content_key=".school",#表示加载特定数据
    text_content=False,#将加载的json对象转换为json字符串
    is_content_key_jq_parsable=True
)
docs = json_loader.load()
print(len(docs))
for doc in docs:
    print(doc)
```

```python
#例3加载json中的education[]中的school和year字段
#例2加载json中edcuation[]中的school字段
from langchain_community.document_loaders import JSONLoader
json_loader = JSONLoader(
    file_path='./asset/4.json',
    jq_schema=".education[]",#表示加载特定数据
    content_key='.school+"\\n\\n"+.year',#表示加载特定数据
    text_content=False,#将加载的json对象转换为json字符串
    is_content_key_jq_parsable=True
)
docs = json_loader.load()
print(len(docs))
for doc in docs:
    print(doc)
```
