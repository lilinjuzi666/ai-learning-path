### 1.单工具的使用
1.1使用ReAct模型


```python
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_core.tools import StructuredTool
from langchain_community.chat_models import ChatTongyi 
from langchain_classic.agents import AgentType,initialize_agent
from langchain_core.tools import Tool
import os
import dotenv
from langchain_community.tools.tavily_search import TavilySearchResults
```

```python
dotenv.load_dotenv()
#配置环境变量
os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY")
os.environ["DASHSCOPE_API_KEY"] = os.getenv("DASHSCOPE_API_KEY")
#获取工具实例
Search = TavilySearchResults(
    max_result=3,

)

#封装工具
#方式1 
# search_tool = StructuredTool.from_function(
#     func=Search.run,
#     name="Search",
#     description="一个天气搜索工具，输入一个字符串，返回搜索结果的字符串",
# )
#方式2
search_tool = Tool(
    name="Search",
    func=Search.run,
    description="一个天气搜索工具，输入一个字符串，返回搜索结果的字符串",
)

#获取模型
llm = ChatTongyi(
    temperature=0.8,
    model="qwen-flash",
)


#获取agent实例
agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION



#获取agentexecutor实例
agent_executor = initialize_agent(
    llm=llm,
    tools=[search_tool],
    agent=agent,
    verbose=True #显示详细的日志信息
)

#调用agent
result = agent_executor.invoke("请帮我搜索一下北京今天的天气")


#处理响应工具
print(result)
```

### 2. 多工具的使用

需求：
- 计算特斯拉当前的股价是多少<br>
- 比去年上涨了百分之几？（调用pythonREPL实例的run方法）

2.1使用ReAct模型

```python
from langchain_experimental.utilities import PythonREPL
dotenv.load_dotenv()
#配置环境变量
os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY")
os.environ["DASHSCOPE_API_KEY"] = os.getenv("DASHSCOPE_API_KEY")
#获取工具实例
Search = TavilySearchResults(
    max_result=3,

)
caculate = PythonREPL()
#封装工具
#工具1 
search_tool = Tool(
    name="Search",
    func=Search.run,
    description="一个天气搜索工具，输入一个字符串，返回搜索结果的字符串",
)

#工具2（计算工具）
caculate_tool = Tool(
    name="caculate",
    func=caculate.run,
    description="用于执行数学计算,例如百分比变化"
)



#获取模型
llm = ChatTongyi(
    temperature=0.8,
    model="qwen-flash",
)


#获取agent实例
agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION



#获取agentexecutor实例
agent_executor = initialize_agent(
    llm=llm,
    tools=[search_tool,caculate_tool],
    agent=agent,
    verbose=True #显示详细的日志信息
)

#调用agent
result = agent_executor.invoke("特斯拉2026年的股价是多少，比2025年上涨了百分之几")


#处理响应工具
print(result)
```

### 3. 自定义函数与工具
计算3的平方，agent调用自定义函数

```python
#自定义函数
def square(expression:str)->str:
    """
    基础数学计算工具，支持加减乘除和幂运算
    Args:
        expression (str): 数字表达式字符串，例如 "3+2*5"
    Returns:
        str: 计算结果字符串或错误信息 
    """
    print(f"\n[工具调用]计算表达式：{expression}")
    print("确认调用")
    return str(eval(expression))#将eval(expression)的结果转换为字符串

#封装工具
math_caculator_tool = Tool(
    name="math_caculator",
    func=square,
    description="用于执行数学计算,例如百分比变化"
)

#获取模型
llm = ChatTongyi(
    temperature=0.8,
    model="qwen-flash",
)


#获取agent实例
agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION



#获取agentexecutor实例
agent_executor = initialize_agent(
    llm=llm,
    agent=agent,
    tools=[math_caculator_tool],
    verbose=True #显示详细的日志信息
)

#调用agent
result = agent_executor.invoke("计算3的平方")


#处理响应工具
print(result)
```
