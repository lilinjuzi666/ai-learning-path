```python
import dotenv
import os
from langchain_community.chat_models import ChatZhipuAI
from langchain.messages import AIMessage,HumanMessage,SystemMessage
from IPython.display import display,Markdown,JSON
```

```python
dotenv.load_dotenv()

os.environ['ZHIPU_API_KEY']=os.getenv('ZHIPUAI_API_KEY')
os.environ['ZHIPU_URL']=os.getenv('ZHIPUAI_URL')

chat=ChatZhipuAI(
    model='glm-4',
)

# messages=[
#     AIMessage(content='你好,有什么可以帮助你的'),
#     SystemMessage(content='你是一个智能助手'),
#     HumanMessage(content='你好')
# ]
response=chat.invoke("什么是大模型")
display(Markdown(response.content))
```

```python
#使用提示词模板
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system","你是世界级的技术文档编写者"),
    ("human","{input}")
])

chain =prompt | chat 
message = chain.invoke({"input":"大模型中的LangChain是什么？"})
display(Markdown(message.content))
```

```python
from langchain_core.output_parsers import JsonOutputParser

prompt = ChatPromptTemplate.from_messages([
    ("system","你是世界级的技术文档编写者"),
    ("human","{input}")
])

output_parser = JsonOutputParser()

chain =prompt | chat | output_parser

result=chain.invoke({"input":"LangChain是什么？用Json格式回复，问题用question表示，答案用answer表示"})
display(JSON(result))
print("*****************")
print(result)

```

```python
#使用向量存储
from langchain_community.document_loaders import WebBaseLoader
import bs4
from zai import ZhipuAiClient

os.environ['ZHIPUAI_API_KEY']=os.getenv('ZHIPUAI_API_KEY')
client = ZhipuAiClient(api_key=os.environ['ZHIPUAI_API_KEY'])
loader =WebBaseLoader(
    web_paths=("https://www.langchain.com.cn/docs/introduction/",),
    bs_kwargs=dict(parse_only=bs4.SoupStrainer(id="__docusaurus"))
)
print(loader)
docs = loader.load()
print(docs)
from langchain_community.embeddings import ZhipuAIEmbeddings

embeddings = ZhipuAIEmbeddings(
    model='embedding-3',
    client=client                         
    )

from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
documents = text_splitter.split_documents(docs)
print (len(documents))
vector = FAISS.from_documents(documents,embeddings)

```

```python
#RAG(Retrieval-Augmented Generation-->>检索增强生成)
from langchain_core.prompts import PromptTemplate


retriever = vector.as_retriever()
retriever.search_kwargs={"k":1}#设置规则
docs = retriever.invoke("什么是LangGraph？")


prompt_template = """
你是一个问答机器人\
你的任务是根据下述给定的一设置信息回答用户问题\
确保你的回复完全依据下述已知信息，不要编造答案\
如果下述已知信息不足以回答用户问题，请直接回复“我无法回复您的问题”\

已知信息；{info}

用户问题；{question}


"""
#得到提示词模板对象
template = PromptTemplate.from_template(prompt_template)

#得到提示词对象
prompt = template.format(info=docs,question="什么是LangGraph？")

response = chat.invoke(prompt)
print(response.content)
```
