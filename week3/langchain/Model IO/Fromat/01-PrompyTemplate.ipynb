## 两种特殊结构的使用（部分提示词模板的使用，组合提示词的使用）

部分提示词模板的使用

```python
#方法1 ：在PeomptTemplate的构造方法或from_template方法中，通过input_variables参数指定变量
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template(
    template="请评价{product}的优缺点,包括{aspect1}和{aspect2}。",
    partial_variables={"product": "智能手机",
                     "aspect1": "电池续航",
                     "aspect2": "拍照质量"},#部分变量填充
)

prompt_1 = template.format()
print(prompt_1)
```

```python
#方法2:调用partial()方法
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template(
    template="请评价{product}的优缺点,包括{aspect1}和{aspect2}。",
    # partial_variables={"product": "智能手机",
    #                  "aspect1": "电池续航",
    #                  "aspect2": "拍照质量"},#部分变量填充
).partial(product="智能手机",aspect1="电池续航", aspect2="拍照质量")
prompt_1 = template.format()
print(prompt_1)
```

组合提示词的使用

```python
from langchain_core.prompts import PromptTemplate


template = (
    PromptTemplate.from_template("Tell me a joke about {topic}")
    + ", make it funny"
    + "\n\nand in {language}"
)


prompt = template.format(topic="sports", language="spanish")
print(prompt)
```

## 给变量辅助的两种方式
format():参数部分给变量赋值，返回一个str类型

invoke():参数部分使用字典给变量赋值，返回一个PromptValue类型----荐

```python
#format
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template(
    template="请评价{product}的优缺点,包括{aspect1}和{aspect2}。",
)

prompt_1 = template.format(product="智能手机", aspect1="电池续航", aspect2="拍照质量")
print(prompt_1)
print(type(prompt_1))
```

```python
#invoke
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template(
    template="请评价{product}的优缺点,包括{aspect1}和{aspect2}。",
)

prompt_1 = template.invoke({"product": "智能手机", "aspect1": "电池续航", "aspect2": "拍照质量"})
print(prompt_1)
print(type(prompt_1))
```

## 结合大模型使用

```python
#导入需要的库
from langchain_community.chat_models.tongyi import ChatTongyi
import dotenv
import os 
#加载环境变量文件中的配置信息
dotenv.load_dotenv()

#获取模型
model = ChatTongyi(
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    model = 'qwen-flash',
    temperature=0.8,
    max_tokens=648,
    )

#生成提示词模板

template = PromptTemplate.from_template(
    template="请评价{product}的优缺点,包括{aspect1}和{aspect2}。",
)

#给模板变量赋值
prompt_1 = template.invoke({"product": "智能手机", "aspect1": "电池续航", "aspect2": "拍照质量"})

#调用模型生成回答
response = model.invoke(prompt_1)
print(response.content)
print(type(response))
```
