### 2. 调用提示词模板的几种方法

invoke() \ format() \ format_messages \ format_prompt

invoke():传入的是字典，返回的是ChatPromptValue对象<br>
format():传入的是变量的值，返回的是字符串<br>
format_messages():传入的是变量的值，返回消息构成的列表<br>
format_prompt():传入的是变量的值，返回的是ChatPromptValue对象


```python
#invoke
from langchain_core.prompts import ChatPromptTemplate


chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system","你是一个ai助手，你的名字叫{name}"),
    ("human","我的问题是{question}")
        ])


response = chat_prompt_template.invoke(input={"name":"小智","question":"1+2+3=？"})
print(response)
print(type(response))
```

```python
#format
from langchain_core.prompts import ChatPromptTemplate


chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system","你是一个ai助手，你的名字叫{name}"),
    ("human","我的问题是{question}")
        ])


response = chat_prompt_template.format(name="小智", question="1+2+3=？")
print(response)
print(type(response))
```

```python
#format_messages
from langchain_core.prompts import ChatPromptTemplate


chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system","你是一个ai助手，你的名字叫{name}"),
    ("human","我的问题是{question}")
        ])


response = chat_prompt_template.format_messages(name="小智", question="1+2+3=？")
print(response)
print(type(response))
```

```python
#format_prompt
from langchain_core.prompts import ChatPromptTemplate


chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system","你是一个ai助手，你的名字叫{name}"),
    ("human","我的问题是{question}")
        ])


response = chat_prompt_template.format_prompt(name="小智", question="1+2+3=？")
print(response)
print(type(response))
```

如何实现ChatPromptValue与list[message],字符串之间的转换

```python
#format_
from langchain_core.prompts import ChatPromptTemplate


chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system","你是一个ai助手，你的名字叫{name}"),
    ("human","我的问题是{question}")
        ])

#将ChatPromptValue对象转换为list[message]
response = chat_prompt_template.invoke({"name":"小智", "question":"1+2+3=？"})
response_message = response.to_messages()
print(response_message)
print(type(response_message))

#将ChatPromptValue对象转换为字符串
response_to_string = response.to_string()
print(response_to_string)
print(type(response_to_string))
```

### 3.更丰富的实例化参数类型

本质上来说，传入的都是消息构成的列表

调用上来说，参数类型都是列表，但是列表的元素是多种多样的

str,dict,元组（最常用）等


```python
#例1
#第一种方式
from langchain_core.prompts import ChatPromptTemplate
chat_prompt_template1 = ChatPromptTemplate(
    messages=[
        ("system","你是一个ai助手，你的名字叫{name}"),
        ("human","我的问题是{question}")
        ],
)

response = chat_prompt_template1.invoke(input={"name":"小智","question":"1+2+3=？"})
print(response)
print(type(response))
print(len(response.messages))


#第二种方式
chat_prompt_template2 = ChatPromptTemplate.from_messages(
    [
    ("system","你是一个ai助手，你的名字叫{name}"),
    ("human","我的问题是{question}")
        ]
)


response = chat_prompt_template2.invoke(input={"name":"小智","question":"1+2+3=？"})
print(response)
print(type(response))
print(len(response.messages))
```

```python
#例2 str
chat_prompt_template3 = ChatPromptTemplate.from_messages(
    ["我的问题是{question}"] #默认的角色是human
)


response = chat_prompt_template3.invoke(input={"name":"小智","question":"1+2+3=？"})
print(response)
print(type(response))
print(len(response.messages))
```

```python
#例3 dict
chat_prompt_template3 = ChatPromptTemplate.from_messages([
    {"role":"system","content":"我是一个ai助手，我的名字叫{name}"},
    {"role":"human","content":"我的问题是{question}"},
]
)


response = chat_prompt_template3.invoke(input={"name":"小智","question":"1+2+3=？"})
print(response)
print(type(response))
print(len(response.messages))
```

```python
#例4 消息类型
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import SystemMessage, HumanMessage
chat_prompt_template3 = ChatPromptTemplate.from_messages([
    SystemMessage(content="我是一个ai助手，我的名字叫{name}"),
    HumanMessage(content="我的问题是{question}")
]
)#不可以再放变量了


response = chat_prompt_template3.invoke({})
print(response)
print(type(response))
print(len(response.messages))
```

```python
#例5Chat提示词模板类型
from langchain_core.prompts import ChatPromptTemplate


chat_prompt_template4 = ChatPromptTemplate.from_messages([
    ("system","你是一个ai助手，你的名字叫{name}"),
        ])
chat_prompt_template5 = ChatPromptTemplate.from_messages([
    ("human","我的问题是{question}")
        ])
chat_prompt_template6 = ChatPromptTemplate.from_messages([
    chat_prompt_template4,
    chat_prompt_template5
        ])



response = chat_prompt_template6.invoke(input={"name":"小智","question":"1+2+3=？"})
print(response)
print(type(response))
print(len(response.messages))
```

```python
#例6 消息提示词模板类型
from langchain_core.prompts import ChatPromptTemplate, HumanMessagePromptTemplate, SystemMessagePromptTemplate

# 创建消息模板

system_message_prompt = SystemMessagePromptTemplate.from_template("你是一个专家{role}")


human_message_prompt = HumanMessagePromptTemplate.from_template("给我解释{concept},用浅显易懂的语言" )
# 组合成聊天提示模板
chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])

# 格式化提示
formatted_messages = chat_prompt.format_messages(
    role="物理学家",
    concept="相对论"
)

print(formatted_messages)
```

### 4. 结合llm

```python
#1.提供大模型
import os 
import dotenv
from  langchain_community.chat_models.tongyi import ChatTongyi
from langchain_core.prompts import ChatPromptTemplate


dotenv.load_dotenv()
chat_model = ChatTongyi(
    model='qwen-flash'
)

#2.通过提示词模板，创建提示词
chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system","你是一个ai助手，你的名字叫{name}"),
    ("human","我的问题是{question}")
        ])


prompt_response = chat_prompt_template.invoke(input={"name":"小智","question":"1+2+3=？"})



#3.将提示词传入大模型，得到结果
response = chat_model.invoke(prompt_response)
print(response.content)

```

### 5.插入消息列表

```python
#使用场景：当ChatPromptTemplate模板中的的消息类型个数不确定时
#例1
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage


chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system","你是一个ai助手，你的名字叫{name}"),
    MessagesPlaceholder(variable_name="messages"),
])

chat_prompt_template.invoke(
    input={"name":"小智",
           "messages":[HumanMessage(content="1+2+3=？"),AIMessage(content="6")]
           
})
```

```python
#例2：存储对话历史记录
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import AIMessage

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    MessagesPlaceholder("history"),
    ("human", "{question}")
])

prompt.format_messages(
    history=[
        HumanMessage(content="1+2*3 = ?"),
        AIMessage(content="1+2*3=7")
    ],
    question="我刚才问题是什么?"
)
```
