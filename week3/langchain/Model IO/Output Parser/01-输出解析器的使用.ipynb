### 1.字符串输出解析器StrOutputParser

```python
#1.获取大模型
import os
import dotenv
from langchain_community.chat_models.tongyi import ChatTongyi
from langchain_core.output_parsers import StrOutputParser

dotenv.load_dotenv()

chat_model =ChatTongyi(
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    model_name="qwen-flash",
)


#2.调用大模型
response = chat_model.invoke("什么是大语言模型")
print(type(response))


#3.如何获取一个字符串的输出结果
#方式1：content 
# print(response.content)
#方式2：StrOutputParser
str_sponse = StrOutputParser().invoke(response)
print(type(str_sponse))
print(str_sponse)
```

### 2. JSON输出解析器JsonOutputParser

```python
#方式1
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import ChatPromptTemplate

dotenv.load_dotenv()

chat_model =ChatTongyi(
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    model_name="qwen-flash",
)


# 定义聊天提示词模板
chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system", "你是一个靠谱的{role}"),
    ("human", "{question}")
])

prompt = chat_prompt_template.invoke(input={"role": "医生", "question": "感冒怎么治疗？问题用q表示，答案用a表示，返回一个json格式的数据"})
response = chat_model.invoke(prompt)
print(response.content)#输出Json字符串

parser = JsonOutputParser()
json_sponse = parser.invoke(response)
print(json_sponse)#将Json字符串转换为python可以直接操作的字典
```

```python
#方式2
parser = JsonOutputParser()
print(parser.get_format_instructions())
#例1
# 引入依赖包
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate

# 初始化语言模型
chat_model = ChatTongyi(model="qwen-flash")

# 定义一个简单的查询字符串
joke_query = "告诉我一个笑话。"

# 定义 Json 解析器
parser = JsonOutputParser()
prompt_template = PromptTemplate.from_template(
    template = "回复用户的查询\n满足的格式为{format_instructions}\n问题为: {question}",
    partial_variables={"format_instructions": parser.get_format_instructions()},
)

prompt = prompt_template.invoke(input={"question":joke_query})
response = chat_model.invoke(prompt)
json_sponse = parser.invoke(response)
print(json_sponse)
```

#### 知识拓展：chain

```python

# 引入依赖包
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import ChatPromptTemplate

dotenv.load_dotenv()

chat_model =ChatTongyi(
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    model_name="qwen-flash",
)


# 定义聊天提示词模板
chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system", "你是一个靠谱的{role}"),
    ("human", "{question}")
])
parser = JsonOutputParser()
#写法1
prompt = chat_prompt_template.invoke(input={"role": "医生", "question": "感冒怎么治疗？问题用q表示，答案用a表示，返回一个json格式的数据"})
response = chat_model.invoke(prompt)
json_sponse = parser.invoke(response)
print(json_sponse)


#写法2 
chain = chat_prompt_template |chat_model |parser
json_result1 = chain.invoke(input={"role": "医生", "question": "感冒怎么治疗？问题用q表示，答案用a表示，返回一个json格式的数据"})
print("-"*50)
print(json_result1)
```

### 4.列表解析器CommaSeparatedListOutputParser

```python
#例1
from langchain_core.output_parsers import CommaSeparatedListOutputParser

# 1. 初始化解析器
output_parser = CommaSeparatedListOutputParser()

# 2. 获取格式化指令
# 这会生成一段类似“Your response should be a list of comma separated values”的英文指令
format_instructions = output_parser.get_format_instructions()
print(format_instructions)

# 3. 模拟 AI 返回的原始字符串
messages = "大象, 猩猩, 狮子"

# 4. 使用解析器进行解析
result = output_parser.parse(messages)

# 5. 打印结果和类型
print(result)
print(type(result))
```

```python
#例2
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import CommaSeparatedListOutputParser


# 1. 初始化语言模型
# 注意：图片中为 "gpt-40-mini"，实际应为 "gpt-4o-mini"
chat_model = ChatTongyi(model="qwen-flash")
# 2. 创建解析器
output_parser = CommaSeparatedListOutputParser()

# 3. 创建 LangChain 提示模板
# 使用 partial_variables 提前注入解析器的格式化指令
chat_prompt = PromptTemplate.from_template(
    "生成5个关于{text}的列表。\n\n{format_instructions}",
    partial_variables={
        "format_instructions": output_parser.get_format_instructions()
    }
)

# 4. 提示模板与输出解析器传递输出（备选方案，代码中已注释）
# chat_prompt = chat_prompt.partial(
#     format_instructions=output_parser.get_format_instructions())

# 将提示和模型合并以进行调用
chain = chat_prompt | chat_model | output_parser
res = chain.invoke({"text": "电影"})
print(res)
print(type(res))
```
