# 3. 关于模型调用的方法的说明
invoke() / stream()<br>
batch():批量的调用<br>
ainvoke() / astream() / abatch()：异步的调用<br>

举例1：invoke()->阻塞式调用（非流式）

```python
from langchain_community.chat_models.tongyi import ChatTongyi
import os
import dotenv

#前提：加载配置文件
dotenv.load_dotenv()


#1.获取对话模型
chat_model = ChatTongyi (
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    model="qwen-flash",
)

messages = [HumanMessage(content="你好，请介绍一下自己")]

#2.调用对话模型，阻塞式输出
response = chat_model.invoke(messages)

print(response)
```

举例1：invoke()->流式

```python
from langchain_community.chat_models.tongyi import ChatTongyi
import os
import dotenv

#前提：加载配置文件
dotenv.load_dotenv()


#1.获取对话模型
chat_model = ChatTongyi (
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    model="qwen-flash",
    streaming=True,#默认为Flase,即阻塞式调用
)

messages = [HumanMessage(content="你好，请介绍一下自己")]

#2.调用对话模型，流式输出
print("开始流式输出：")
for chunk in chat_model.stream(messages):
    print(chunk.content,end="",flush=True)#end=""表示不换行，flush=True表示刷新缓冲区，立即输出
print("\nend")
```

例3：使用batch测试批量调用

```python
from langchain_community.chat_models.tongyi import ChatTongyi

chat_model_qwq = ChatTongyi(
    api_key=os.getenv("DASHSCOPE_API_KEY"),
    model='qwen-flash'
)
messages1 = [HumanMessage(content="请帮我介绍一下什么是机器学习")]
messages2 = [HumanMessage(content="请帮我介绍一下什么是AIGC")]
messages3 = [HumanMessage(content="请帮我介绍一下什么是大模型技术")]
messages = [messages1, messages2, messages3]
response = chat_model_qwq.batch(messages)

for i,resp in enumerate(response):
    print(f"第{i+1}个请求的响应是：{resp.content}")
```

例4：关于同步和异步方法的调用

```python
#同步
import time

def call_model():
    #模拟同步api调用
    print("开始调用模型...")
    time.sleep(5)#模型调用等待，单位：秒
    print("模型调用结束")

def perform_other_tasks():
    #模拟执行其他任务
    for i in range(5):#range(5)生成一个从到4的整数序列
        print(f"执行其他任务{i+1}")
        time.sleep(1)

def main():
    start_time = time.time()#计时起点
    call_model()
    perform_other_tasks()
    end_time = time.time()#计时终点
    total_time = end_time - start_time
    return (f"总共耗时{total_time}秒")

main_time = main()
print(main_time)

```

```python
#异步
import time
import asyncio #asyncio是Python的异步编程库

async def async_call(llm):
    await asyncio.sleep(5)  #暂停5s时可以先执行其他操作
    print("异步调用完成")

async def perform_other_tasks():
    await asyncio.sleep(5)
    print("其他任务完成")

async def run_async_tasks():
    start_time = time.time()
    await asyncio.gather(
        async_call(None),
        perform_other_tasks()
    )
    end_time = time.time()
    return (f"总共耗时{end_time - start_time}秒")


result = await run_async_tasks()
print(result)
```
