```python
#传统写法
def add(a,b):
    return a+b
```

```python
#工程化写法
def add(a:int,b:int)->int:
    """
    计算两个整数的和
    :param a: 第一个整数
    :param b: 第二个整数
    :return: 两个整数的和
    """
    return a+b


result=add(10,20)
print(f"结果是：{result}")
```
