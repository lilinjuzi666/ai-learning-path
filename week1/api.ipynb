```python
import os
from dotenv import load_dotenv

load_dotenv('api.env')#读取.env内容并保存到系统的环境变量中

api_key = os.getenv("MY_API_KEY")

if api_key:
    print(f"成功读取到Key:{api_key}")
else:
    print("未能读取到Key,请检查.env文件是否存在，并且包含正确的Key")
```

数据校验

```python
from pydantic import BaseModel,ValidationError,EmailStr

class User(BaseModel):#BaseModel是pydantic提供的一个基类，所有需要校验的模型都需要继承这个类
    name:str
    age:int
    email:EmailStr

valid_data={
    "name":"CodeGeex",
    "age":3,
    "email":"codegeex@example.com"

    }

invalid_data={
    "name":"BadUser",
    "age":"twenty",
    "email":"not-an-email"
    }

try:
    user=User(**valid_data)
    print("校验成功：",user)
except ValidationError as e:
    print("校验失败：",e)
try:
    bad_user=User(**invalid_data)
    print("校验成功：",bad_user)
except ValidationError as e:
    print("校验失败：",e)
```
