简单的天气查询脚本

```python
import requests
import json
from dotenv import load_dotenv
import os
from pydantic import BaseModel, ValidationError
```

读取承德经纬度

```python
load_dotenv('weather.env')#读取.env内容并保存到系统的环境变量中
jwd=os.getenv("chengde_lati_longi")
print(f"承德的经纬度是：{jwd}")
```

调用Open-Meteo API获取天气数据

```python
tem=requests.get("https://api.open-meteo.com/v1/forecast?latitude=40.57&longitude=117.57&current_weather=true").json()
print(tem)
print(f"承德当前的温度是：{tem['current_weather']['temperature']}°C")
```

pydantic数据校验模型

```python
#校验数据格式
class CurrentWeather(BaseModel):
    temperature: float
class WeatherResponse(BaseModel):
    current_weather: CurrentWeather

try:
    weather=CurrentWeather(**tem['current_weather'])#只验证current_weather部分
    print(f"校验成功:当前温度是{tem['current_weather']['temperature']}°C")
except ValidationError as e:
    print("校验失败:", e)
```
