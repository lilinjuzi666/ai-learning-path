AI英语作文润色师

```python
import os
from zhipuai import ZhipuAI
from IPython.display import display,Markdown
from dotenv import load_dotenv
```

```python

def get_completion(messages):
            response = client.chat.completions.create(
                model="glm-4",
                messages=messages,
            )
            return response.choices[0].message.content#链式调用

def main ():
    _=load_dotenv('.env')
    client=ZhipuAI(api_key=os.getenv('API_KEY'))
    system_prompt="""

        你是一位资深的雅思英语写作考官
        下面是一篇学生写的英语作文，请你：
        1. 纠正并指出所有的语法错误。
        2. 提供一个更地道的改写版本。
        3. 用中文简要评价这篇作文的优缺点。
        4. 在回答问题时不要啰嗦
        '''{user_input}'''
        """  
    history=[]
    history.append({"role":"user","content":system_prompt})
    print(f"你好，我是AI英语作文润色师，请输入需要润色的英语作文")
    while True:
        user_input = input("请输入需要润色的英语作文：")
        if user_input.lower() == "exit":
            print("感谢使用，再见！")
            break
        print(f"用户：{user_input}")
        history.append({"role":"user","content":user_input})
        response = get_completion(history)
        ai_response = display(Markdown(response))
        history.append({"role":"assistant","content":response})


    
```

```python
if __name__ == "__main__":
    main()
```
