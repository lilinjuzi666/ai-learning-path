```python
from zhipuai import ZhipuAI 
import os
from dotenv import load_dotenv,find_dotenv

_=load_dotenv(find_dotenv())
ZhipuAI.api_key=os.getenv("ZHIPUAI_API_KEY")
```

```python
client = ZhipuAI(api_key=os.getenv("ZHIPUAI_API_KEY"))


def get_completion(prompt, model="glm-4"):
    response = client.chat.completions.create(  # 新的调用方式
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )
    
    return response.choices[0].message.content  # 访问方式略有变化

    
```

```python
text = f"""
您应该通过提供尽可能清晰和具体的指令来表达您希望模型执行的任务。\
这将引导模型生成期望的输出，并减少收到无关或不正确响应的机会。\
不要将编写清晰的提示词与编写简短的提示词混淆。\
在许多情况下，较长的提示词能为模型提供更多的清晰度和上下文，从而产生更详细和相关的输出。\
"""
prompt = f"""
将三个反引号分隔的文本总结为一句话。\
'''{text}'''
"""
response = get_completion(prompt)
print(response)
```
